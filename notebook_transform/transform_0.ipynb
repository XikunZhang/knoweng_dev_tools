# Transpose a spreadsheet and rewrite it with a "_T" appended to the name.


```python
import os
import numpy as np
import pandas as pd
import knpackage.toolbox as kn
```

#### Get two dataframes, match up the sample names and output two dataframes with those matching sample names.

```python
def read_copy_transpose_write_df(data_directory, spreadsheet_file):
    #data_directory, spreadsheet_file = get_run_directory_and_file(sys.argv)
    full_file_name = os.path.join(data_directory, spreadsheet_file)
    spreadsheet_df_0 = pd.read_csv(full_file_name, sep='\t', index_col=0, header=0)
    spreadsheet_df = spreadsheet_df_0.transpose()
    output_file_name = os.path.join(data_directory, spreadsheet_file[0:-4] + '_T.tsv')
    spreadsheet_df.to_csv(output_file_name, sep='\t', index=True, header=True)
    return spreadsheet_df_0, spreadsheet_df
```

##### Create the test data; split a samples x phenotypes dataframe into two samples-overlapping  dataframes.

```python
data_source_directory = './transform_data'
os.listdir(data_source_directory)
```

```python
spreadsheet_df_0, spreadsheet_df = read_copy_transpose_write_df(data_source_directory, 'UCEC_phenotype.txt')
spreadsheet_df_0
```
